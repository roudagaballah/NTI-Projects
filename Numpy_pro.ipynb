```python
import numpy as np
```

```python
fitness_data = np.array([[5000, 200, 6],[7000, 250, 7],[6500, 230, 6],[8000, 300, 8],[9000, 
320, 7],[7500, 270, 6],[12000, 400, 5]])  
```

================== Explore the array ===============

```python
# Number of dimensions 
np.ndim(fitness_data)
```

```python
#Shape of the array  
np.shape(fitness_data)
```

```python
 # Total number of elements  
np.size(fitness_data)
```

======================= Slice the data============

```python
# All step values 
steps = fitness_data[ : , 0]
print(steps)
```

```python
# All calorie values 
calorie = fitness_data[: , 1]
print(calorie)
```

```python
# All sleep values
sleep_hours = fitness_data[:, 2]
print(sleep_hours)
```

```python
# First 3 days 
First_3_days = fitness_data[0:3 , :] 
print(First_3_days)
```

```python
# Steps and calories only 
col0_col1 = fitness_data[: , 0:2]
print(col0_col1)
```

======================= Reshape the data =============

```python
# Reshape the steps array into a column (7,1) 
np.reshape( steps,(7,1))
```

```python
# Reshape the full dataset into shape (3,7) 
np.reshape(fitness_data,(3,7))
```

```python
print(steps>8000)
```

======================= Filter the data =================

```python
fitness_data[steps>8000]
```

```python
fitness_data[sleep_hours < 6]
```

```python
fitness_data[calorie>300]
```
